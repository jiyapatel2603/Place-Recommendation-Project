```python
main_data= data.drop(['Name', 'Date', 'Raw_Review'], axis=1)
```

```python
city_list = main_data['City'].unique().tolist()
city_list
```

```python
city_dropdown = widgets.Dropdown(
    options=city_list,
    value=city_list[0],
    description='City:',)
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer, PorterStemmer
```

```python
# nltk.download('stopwords')
# nltk.download('wordnet')
# nltk.download('omw-1.4')
```

```python
# Stemming/Lemmatizing

stop_words = set(stopwords.words('english'))
main_data['Review'] = main_data['Review'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))


stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()
main_data['Review'] = main_data['Review'].apply(lambda x: ' '.join([stemmer.stem(word) for word in x.split()]))
main_data['Review'] = main_data['Review'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
```

```python
# Convert text to lowercase
main_data['Review'] = main_data['Review'].str.lower()
```

```python
from textblob import TextBlob

# Calculate sentiment scores

main_data['Sentiment_Score'] = main_data['Review'].apply(lambda x: TextBlob(x).sentiment.polarity)
```

```python
# Normalize sentiment scores

main_data['Normalized_Sentiment_Score'] = (main_data['Sentiment_Score'] - main_data['Sentiment_Score'].min()) / (main_data['Sentiment_Score'].max() - main_data['Sentiment_Score'].min())
```

```python
# Calculate utility scores

main_data_utility = main_data.groupby(['City', 'Place']).agg({'Normalized_Sentiment_Score': 'mean', 'Review': 'count'}).reset_index()
main_data_utility['Utility_Score'] = main_data_utility['Normalized_Sentiment_Score'] * main_data_utility['Review']
```

```python
# Sort places by utility score

main_data_sorted = main_data_utility.sort_values(by=['City', 'Utility_Score'], ascending=[True, False]).reset_index(drop=True)
```

```python
display(city_dropdown)
```

```python
# Recommend top 10 places and their avg rating
# Specify the city for which you want to make recommendations

city = city_dropdown.value
top_places = main_data_sorted[main_data_sorted['City'] == city].head(10)['Place'].tolist()

top_places_df = pd.DataFrame(top_places, columns=['Place'])
avg_ratings_df = main_data.groupby('Place')['Rating'].mean()
final_df = top_places_df.merge(avg_ratings_df, on='Place')
final_df['Rating'] = round(final_df.Rating, 2) 
final_df.index = final_df.index + 1

#print final output
print(f'Top 10 places to visit in {city}:')
final_df
```
